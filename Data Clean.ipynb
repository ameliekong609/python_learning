```python
raw_accident_3yr = pd.concat([raw_accident_2018,raw_accident_2019,raw_accident_2020])
raw_accident_3yr.shape
```

```python
raw_accident_3yr.columns
```

```python
###Mapping with dictionary
mapping=pd.read_csv("Accident_Severity.csv")
mapping_accident_severity = dict(mapping[['code', 'label']].values)
raw_accident_3yr['accident_severity1'] = raw_accident_3yr['accident_severity'].map(mapping_accident_severity)

mapping=pd.read_csv("Day_of_Week.csv")
mapping_day_of_week = dict(mapping[['code', 'label']].values)
raw_accident_3yr['day_of_week1'] = raw_accident_3yr['day_of_week'].map(mapping_day_of_week)

mapping=pd.read_csv("Junction_Control.csv")
mapping_junction_control = dict(mapping[['code', 'label']].values)
raw_accident_3yr['junction_control1'] = raw_accident_3yr['junction_control'].map(mapping_junction_control)

mapping=pd.read_csv("Junction_Detail.csv")
mapping_junction_detail = dict(mapping[['code', 'label']].values)
raw_accident_3yr['junction_detail1'] = raw_accident_3yr['junction_detail'].map(mapping_junction_detail)

mapping=pd.read_csv("Light_Conditions.csv")
mapping_light_condition = dict(mapping[['code', 'label']].values)
raw_accident_3yr['light_conditions1'] = raw_accident_3yr['light_conditions'].map(mapping_light_condition)

mapping=pd.read_csv("Local_Authority_District.csv")
mapping_local_authority_district = dict(mapping[['code', 'label']].values)
raw_accident_3yr['local_authority_district1'] = raw_accident_3yr['local_authority_district'].map(mapping_local_authority_district)

mapping=pd.read_csv("Local_Authority_Highway.csv")
mapping_Local_Authority_Highway = dict(mapping[['Code', 'Label']].values)
raw_accident_3yr['local_authority_highway1'] = raw_accident_3yr['local_authority_highway'].map(mapping_Local_Authority_Highway)


mapping=pd.read_csv("Ped_Cross_Human.csv")
mapping_Ped_Cross_Human = dict(mapping[['code', 'label']].values)
raw_accident_3yr['pedestrian_crossing_human_control1'] = raw_accident_3yr['pedestrian_crossing_human_control'].map(mapping_Ped_Cross_Human)


mapping=pd.read_csv("Ped_Cross_Physical.csv")
mapping_Ped_Cross_Physical = dict(mapping[['code', 'label']].values)
raw_accident_3yr['pedestrian_crossing_physical_facilities1'] = raw_accident_3yr['pedestrian_crossing_physical_facilities'].map(mapping_Ped_Cross_Physical)

mapping=pd.read_csv("Police_Force.csv")
mapping_Police_Force = dict(mapping[['code', 'label']].values)
raw_accident_3yr['police_force1'] = raw_accident_3yr['police_force'].map(mapping_Police_Force)

mapping=pd.read_csv("Police_Officer_Attend.csv")
mapping_Police_Force = dict(mapping[['code', 'label']].values)
raw_accident_3yr['did_police_officer_attend_scene_of_accident1'] = raw_accident_3yr['did_police_officer_attend_scene_of_accident'].map(mapping_Police_Force)

mapping=pd.read_csv("Road_Class.csv")
mapping_Road_Class = dict(mapping[['code', 'label']].values)
raw_accident_3yr['first_road_class1'] = raw_accident_3yr['first_road_class'].map(mapping_Road_Class)

mapping=pd.read_csv("Road_Type.csv")
mapping_Road_Type = dict(mapping[['code', 'label']].values)
raw_accident_3yr['road_type1'] = raw_accident_3yr['road_type'].map(mapping_Road_Type)

mapping=pd.read_csv("Urban_Rural.csv")
mapping_Urban_Rural = dict(mapping[['code', 'label']].values)
raw_accident_3yr['urban_or_rural_area1'] = raw_accident_3yr['urban_or_rural_area'].map(mapping_Urban_Rural)
```
